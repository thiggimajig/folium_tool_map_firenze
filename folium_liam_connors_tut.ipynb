```python
import pandas as pd
import folium
```

creating a dataframe from our dataset csv file 

```python
listings = "../thesis_app/Out_CSV/florence_listings_jan_27_cleaned_trimmed500_2022_02_15_21_51_51.csv"
prelim_listings_location = pd.read_csv(listings)
```

shorten dataframe even more so doensn't take so long

```python
listings_location = prelim_listings_location.loc[:50, :]
```

checking the new dataframe

```python
listings_location.head()
```

keeping only the columns we want

```python
listings_location = listings_location[["name", "host_total_listings_count", "price", "listing_url", "host_name", "longitude", "latitude"]]
```

creating the map

```python
#m = folium.Map(location=[45.5236, -122.6750])
map = folium.Map(location=[listings_location.latitude.mean(),listings_location.longitude.mean()], zoom_start=12, control_scale=True) #defaults to OSM can adjust tiles but need attribution can't figure that out
```

show map

```python
map.save("firenze_folium_map.html") 
map
```

Adding points to the map, by loopin over all listings rows in dataframe and pulling out important location info and popup information like name from other columns not long lat

```python
for index, location_info in listings_location.iterrows():
    folium.Marker([location_info["latitude"],location_info["longitude"]], popup=[location_info["name"],location_info["price"],location_info["host_total_listings_count"]]).add_to(map)
#why was location_info not an issue? never declared it before? 
```

```python
map.save("firenze_folium_map.html")
map
```

other fun things to customize

```python
#can pass escaped html into popup 
#parse_html=True 
#popup (string or folium.Popup, default None) â€“ Label for the Marker; either an escaped HTML string to initialize folium.Popup or a folium.Popup instance.
#this isn't html but it's prettier... 
for index, location_info in listings_location.iterrows():
    folium.Marker([location_info["latitude"],location_info["longitude"]], popup="Total Listings Count: " + str(location_info["host_total_listings_count"])).add_to(map)
```

```python
map
```

```python
#tooltip (str or folium.Tooltip, default None) â€“ Display a text when hovering over the object.
for index, location_info in listings_location.iterrows():
    folium.Marker([location_info["latitude"],location_info["longitude"]], popup=[location_info["name"],location_info["price"],location_info["host_total_listings_count"]], tooltip=location_info["name"]).add_to(map)
```

```python
map.save("firenze_folium_map.html")
map
```
